## Distribuciones $X^{(c)}$

### $X^{PPD} \sim \mathcal{LN}(14.8219, 0.9396)$

```r
  # ParÃ¡metros de la lognormal
  mu <- 14.8219
  sigma <- 0.9396

  # Crear el grÃ¡fico
  grafico <- ggplot() +
    # Histograma de los datos
    geom_histogram(data = data.frame(x = severidad_ppd),
                   aes(x = x, y = after_stat(density)),
                   bins = 100, alpha = 0.7, fill = "lightblue", color = "white") +
    # Curva de la lognormal ajustada
    stat_function(fun = dlnorm, args = list(meanlog = mu, sdlog = sigma),
                  color = "red", size = 1.2) +
    labs(title = "Ajuste Lognormal a Datos de Severidad PPD",
         x = "Severidad", y = "Densidad") +
    theme_minimal()

  grafico

  # Guardar en /images
  ggsave("../images/ajuste_lognormal_ppd.png", grafico, width = 10, height = 6, dpi = 300)
```
