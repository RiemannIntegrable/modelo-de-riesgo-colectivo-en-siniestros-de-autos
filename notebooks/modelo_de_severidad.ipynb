## Librerias

```r
library(ggplot2)
```

## Data

```r
ppd <- read.csv("../data/processed/severidad_ppd.csv")
pph <- read.csv("../data/processed/severidad_pph.csv")
pth <- read.csv("../data/processed/severidad_pth.csv")
rc <- read.csv("../data/processed/severidad_rc.csv")
```

## Graficas

### PPD

```r
  p <- ggplot(data.frame(x = ppd$x), aes(x = x)) +
    geom_density(fill = "steelblue", color = "darkblue", alpha = 0.7, size = 1) +
    labs(title = "DistribuciÃ³n de la Severidad de ppd",
         x = "Monto del Siniestro",
         y = "Densidad") +
    theme_minimal() +
    theme(panel.background = element_rect(fill = "white"),
          plot.background = element_rect(fill = "white"),
          panel.grid.minor = element_blank(),
          plot.title = element_text(hjust = 0.5, size = 14, face = "bold"),
          axis.title = element_text(size = 12))

  # Mostrar con dimensiones 16:9
  options(repr.plot.width = 12, repr.plot.height = 4)
  p

  ggsave("../images/distribucion_severidad_ppd.png", p, width = 16, height = 9, dpi = 300, bg = "white")
```
