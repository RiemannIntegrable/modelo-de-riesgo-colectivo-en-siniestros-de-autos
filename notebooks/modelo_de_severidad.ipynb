```r
## Gráficas de distribuciones ajustadas

### Obtener distribuciones recomendadas

resultado_ppd <- Estimadores_severidad(severidad_ppd)
resultado_pph <- Estimadores_severidad(severidad_pph)
resultado_pth <- Estimadores_severidad(severidad_pth)
resultado_rc <- Estimadores_severidad(severidad_rc)

# Extraer información de cada distribución recomendada
dist_ppd <- resultado_ppd$recomendacion$distribucion
params_ppd <- resultado_ppd$ajustes[[dist_ppd]]$parametros

dist_pph <- resultado_pph$recomendacion$distribucion
params_pph <- resultado_pph$ajustes[[dist_pph]]$parametros

dist_pth <- resultado_pth$recomendacion$distribucion
params_pth <- resultado_pth$ajustes[[dist_pth]]$parametros

dist_rc <- resultado_rc$recomendacion$distribucion
params_rc <- resultado_rc$ajustes[[dist_rc]]$parametros

print(paste("PPD:", dist_ppd))
print(paste("PPH:", dist_pph))
print(paste("PTH:", dist_pth))
print(paste("RC:", dist_rc))
```

```r
### Función para crear gráficas de distribuciones ajustadas

crear_grafica_ajuste <- function(datos, dist_name, params, titulo) {
  library(ggplot2)
  
  # Crear histograma de los datos
  datos_df <- data.frame(x = datos)
  
  # Calcular valores teóricos de la distribución
  x_seq <- seq(min(datos), max(datos), length.out = 1000)
  
  if (dist_name == "weibull") {
    y_teorico <- dweibull(x_seq, shape = params$shape, scale = params$scale)
  } else if (dist_name == "gamma") {
    y_teorico <- dgamma(x_seq, shape = params$shape, rate = params$rate)
  } else if (dist_name == "lognormal") {
    y_teorico <- dlnorm(x_seq, meanlog = params$meanlog, sdlog = params$sdlog)
  } else if (dist_name == "normal") {
    y_teorico <- dnorm(x_seq, mean = params$mean, sd = params$sd)
  }
  
  dist_df <- data.frame(x = x_seq, y = y_teorico)
  
  # Crear gráfica
  p <- ggplot() +
    geom_histogram(data = datos_df, aes(x = x, y = ..density..), 
                   bins = 200, alpha = 0.7, fill = "lightblue", color = "darkblue") +
    geom_line(data = dist_df, aes(x = x, y = y), 
              color = "red", size = 1.2, alpha = 0.8) +
    labs(title = paste("Ajuste", stringr::str_to_title(dist_name), "para", titulo),
         x = "Monto del Siniestro",
         y = "Densidad") +
    theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5, size = 10, face = "bold"),
          axis.title = element_text(size = 9),
          axis.text = element_text(size = 8))
  
  return(p)
}
```

```r
### Crear gráficas individuales

library(gridExtra)
library(grid)

# Crear las 4 gráficas
p1 <- crear_grafica_ajuste(severidad_ppd, dist_ppd, params_ppd, "PPD")
p2 <- crear_grafica_ajuste(severidad_pph, dist_pph, params_pph, "PPH")
p3 <- crear_grafica_ajuste(severidad_pth, dist_pth, params_pth, "PTH")
p4 <- crear_grafica_ajuste(severidad_rc, dist_rc, params_rc, "RC")

# Crear malla 2x2 con dimensiones 16:9
options(repr.plot.width = 16, repr.plot.height = 9)

# Título general
titulo_general <- textGrob("Distribuciones de Severidad Ajustadas por Cobertura", 
                          gp = gpar(fontsize = 16, fontface = "bold"))

# Crear la malla
grid_plot <- grid.arrange(
  p1, p2, p3, p4,
  ncol = 2, nrow = 2,
  top = titulo_general
)

# Guardar la imagen
ggsave("../images/ajuste_distribuciones_severidad.png", grid_plot, 
       width = 16, height = 9, dpi = 300, bg = "white")

print("Gráficas creadas y guardadas exitosamente")
```

```r
## Discretización de distribuciones de severidad

### Cargar función de discretización y parámetros

source("../src/utils/Discretizar_severidad.R")

# Parámetros de discretización
paso_discretizacion <- 10000
limite_superior <- 500000000  # 500 millones como límite superior

cat("Distribuciones y parámetros para discretización:\n")
cat(sprintf("PPD: %s (%s)\n", stringr::str_to_title(dist_ppd), resultado_ppd$recomendacion$parametros))
cat(sprintf("PPH: %s (%s)\n", stringr::str_to_title(dist_pph), resultado_pph$recomendacion$parametros))
cat(sprintf("PTH: %s (%s)\n", stringr::str_to_title(dist_pth), resultado_pth$recomendacion$parametros))
cat(sprintf("RC: %s (%s)\n", stringr::str_to_title(dist_rc), resultado_rc$recomendacion$parametros))

print("Función de discretización cargada exitosamente")
```

```r
### Discretización para cada cobertura

# PPD - Gamma
severidad_discreta_ppd <- Discretizar_severidad(
  distribucion = dist_ppd,
  parametros = params_ppd,
  paso = paso_discretizacion,
  limite = limite_superior
)

cat("PPD discretizada:\n")
cat(sprintf("Distribución: %s\n", stringr::str_to_title(dist_ppd)))
cat(sprintf("Parámetros: %s\n", resultado_ppd$recomendacion$parametros))
cat(sprintf("Número de puntos: %d\n", severidad_discreta_ppd$numero_puntos))
cat(sprintf("Suma de probabilidades: %.6f\n", severidad_discreta_ppd$suma_probabilidades))

# PPH - Gamma
severidad_discreta_pph <- Discretizar_severidad(
  distribucion = dist_pph,
  parametros = params_pph,
  paso = paso_discretizacion,
  limite = limite_superior
)

cat("\nPPH discretizada:\n")
cat(sprintf("Distribución: %s\n", stringr::str_to_title(dist_pph)))
cat(sprintf("Parámetros: %s\n", resultado_pph$recomendacion$parametros))
cat(sprintf("Número de puntos: %d\n", severidad_discreta_pph$numero_puntos))
cat(sprintf("Suma de probabilidades: %.6f\n", severidad_discreta_pph$suma_probabilidades))

# PTH - Gamma
severidad_discreta_pth <- Discretizar_severidad(
  distribucion = dist_pth,
  parametros = params_pth,
  paso = paso_discretizacion,
  limite = limite_superior
)

cat("\nPTH discretizada:\n")
cat(sprintf("Distribución: %s\n", stringr::str_to_title(dist_pth)))
cat(sprintf("Parámetros: %s\n", resultado_pth$recomendacion$parametros))
cat(sprintf("Número de puntos: %d\n", severidad_discreta_pth$numero_puntos))
cat(sprintf("Suma de probabilidades: %.6f\n", severidad_discreta_pth$suma_probabilidades))

# RC - Lognormal
severidad_discreta_rc <- Discretizar_severidad(
  distribucion = dist_rc,
  parametros = params_rc,
  paso = paso_discretizacion,
  limite = limite_superior
)

cat("\nRC discretizada:\n")
cat(sprintf("Distribución: %s\n", stringr::str_to_title(dist_rc)))
cat(sprintf("Parámetros: %s\n", resultado_rc$recomendacion$parametros))
cat(sprintf("Número de puntos: %d\n", severidad_discreta_rc$numero_puntos))
cat(sprintf("Suma de probabilidades: %.6f\n", severidad_discreta_rc$suma_probabilidades))

print("Discretización completada para todas las coberturas")
```

```r
### Exportar vectores de probabilidad discretizados

# Extraer vectores de probabilidades para usar en Panjer
prob_severidad_ppd <- severidad_discreta_ppd$probabilidades
prob_severidad_pph <- severidad_discreta_pph$probabilidades
prob_severidad_pth <- severidad_discreta_pth$probabilidades
prob_severidad_rc <- severidad_discreta_rc$probabilidades

# Crear metadata para cada cobertura
metadata_ppd <- data.frame(
  cobertura = "PPD",
  distribucion = dist_ppd,
  parametros = resultado_ppd$recomendacion$parametros,
  paso = paso_discretizacion,
  limite = limite_superior,
  numero_puntos = severidad_discreta_ppd$numero_puntos,
  suma_probabilidades = severidad_discreta_ppd$suma_probabilidades
)

metadata_pph <- data.frame(
  cobertura = "PPH",
  distribucion = dist_pph,
  parametros = resultado_pph$recomendacion$parametros,
  paso = paso_discretizacion,
  limite = limite_superior,
  numero_puntos = severidad_discreta_pph$numero_puntos,
  suma_probabilidades = severidad_discreta_pph$suma_probabilidades
)

metadata_pth <- data.frame(
  cobertura = "PTH",
  distribucion = dist_pth,
  parametros = resultado_pth$recomendacion$parametros,
  paso = paso_discretizacion,
  limite = limite_superior,
  numero_puntos = severidad_discreta_pth$numero_puntos,
  suma_probabilidades = severidad_discreta_pth$suma_probabilidades
)

metadata_rc <- data.frame(
  cobertura = "RC",
  distribucion = dist_rc,
  parametros = resultado_rc$recomendacion$parametros,
  paso = paso_discretizacion,
  limite = limite_superior,
  numero_puntos = severidad_discreta_rc$numero_puntos,
  suma_probabilidades = severidad_discreta_rc$suma_probabilidades
)

# Exportar archivos completos con valores y probabilidades
write.csv(data.frame(
  valores = severidad_discreta_ppd$valores,
  probabilidades = prob_severidad_ppd
), "../data/processed/severidad_discreta_ppd.csv", row.names = FALSE)

write.csv(data.frame(
  valores = severidad_discreta_pph$valores,
  probabilidades = prob_severidad_pph
), "../data/processed/severidad_discreta_pph.csv", row.names = FALSE)

write.csv(data.frame(
  valores = severidad_discreta_pth$valores,
  probabilidades = prob_severidad_pth
), "../data/processed/severidad_discreta_pth.csv", row.names = FALSE)

write.csv(data.frame(
  valores = severidad_discreta_rc$valores,
  probabilidades = prob_severidad_rc
), "../data/processed/severidad_discreta_rc.csv", row.names = FALSE)

# Exportar solo vectores de probabilidades para uso directo en Panjer
write.csv(data.frame(probabilidades = prob_severidad_ppd), 
          "../data/processed/prob_severidad_ppd.csv", row.names = FALSE)
write.csv(data.frame(probabilidades = prob_severidad_pph), 
          "../data/processed/prob_severidad_pph.csv", row.names = FALSE)
write.csv(data.frame(probabilidades = prob_severidad_pth), 
          "../data/processed/prob_severidad_pth.csv", row.names = FALSE)
write.csv(data.frame(probabilidades = prob_severidad_rc), 
          "../data/processed/prob_severidad_rc.csv", row.names = FALSE)

# Exportar metadata
metadata_completa <- rbind(metadata_ppd, metadata_pph, metadata_pth, metadata_rc)
write.csv(metadata_completa, "../data/processed/severidad_discreta_metadata.csv", row.names = FALSE)

print("Vectores de severidad discretizados exportados exitosamente")
```
